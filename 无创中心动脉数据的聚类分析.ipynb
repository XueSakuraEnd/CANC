# 一,数据处理
* 缺省值处理(**Default handle**):  
所用数据中,目标特征发现有几例缺省值,数量较少,故剔除
>   
* 数据洗牌(**Shuffle**):  
消除数据顺序的干扰,增加随机性,使数据的分布更符合原始分布,降低过拟合性
*  数据标准化(**Normalize**):    
避免因为特征量纲的差异性导致模型学习过程中出现较大的特征偏移使得拟合效果变差,对特征进行归一化处理,将各个指标转化为均值为0,标准差为1的标准化数据
>  

# 二,聚类过程  
* 确认最优聚类数目K:  
使用卡林斯基-哈拉巴斯指数进行对聚类效果的评估,当簇内差异越小,簇间差异越大时,该指数就越大,根据分析不同k值的指数,得出突变点k=2聚类效果最好
>
>
* 簇中心的选取:
使用了两种簇中心初始化方法,一种是随机初始化,另一种是kmeans++的初始化方法,效果相同
>
>
>
* 聚类的算法过程:  
使用欧式距离确定簇中心  
* 终止:  
迭代终止或者这一轮的簇中心与上一轮簇中心一致则停止迭代
